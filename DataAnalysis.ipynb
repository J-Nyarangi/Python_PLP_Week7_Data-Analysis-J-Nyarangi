```python
# Basic statistics of numerical columns
print("\nBasic Statistics of Numerical Columns:")
print(df.describe())
```

```python
# Group by species and calculate the mean of numerical columns
mean_by_species = df.groupby('species').mean()
print("\nMean of numerical columns grouped by species:")
print(mean_by_species)
```

```python
# Bar chart - Average petal length per species
plt.figure(figsize=(8, 5))
mean_by_species['petal length (cm)'].plot(kind='bar', color=['skyblue', 'orange', 'green'], edgecolor='black')
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.xticks(rotation=0)
plt.show()
```

```python
# Histogram - Distribution of sepal width
plt.figure(figsize=(8, 5))
plt.hist(df['sepal width (cm)'], bins=20, color='purple', edgecolor='black')
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot - Sepal length vs. petal length
plt.figure(figsize=(8, 5))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=df)
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```
